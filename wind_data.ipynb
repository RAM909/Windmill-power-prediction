```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# Load the dataset
data = pd.read_csv("par1final.csv")

# Select features (independent variables) and target variable
X = data[['Air temperature | (°C)', 'Pressure | (atm)', 'Wind speed | (m/s)', 'year', 'month', 'day', 'hour']]
y = data['Power generated by system | (MW)']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)

# Train the model
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# Ask user for input data
temp = float(input("Enter air temperature (°C): "))
pressure = float(input("Enter pressure (atm): "))
windspeed = float(input("Enter wind speed (m/s): "))
year = int(input("Enter year: "))
month = int(input("Enter month: "))
day = int(input("Enter day: "))
hour = int(input("Enter hour: "))

# Create a DataFrame for new data
new_data = pd.DataFrame({
    'Air temperature | (°C)': [temp],
    'Pressure | (atm)': [pressure],
    'Wind speed | (m/s)': [windspeed],
    'year': [year],  
    'month': [month],     
    'day': [day],       
    'hour': [hour],       
})

# Predict power generated by the system
predicted_power = model.predict(new_data)

# Display the predicted power
print("Predicted power generated by the system (MW):", predicted_power[0])
```
